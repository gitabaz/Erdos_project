# 2018 election

The main goal of this work is to identify the most important issues leading up into the election.

Our strategy is to
1. Gather all the politics articles from 2018.
2. Build a topic model for the year.
3. Track each of the topics throughout the year and study any changes as the election approaches.

# Random forest classifier

We will build a random forest classifier trained on the labeled section and use that to create labels for our unlabeled dataset.

We have to do a bit more preprocessing for the classifier:
- Make each section lower case so politics and Politics are the same
- One-hot encode the section so that 0 = not politics and 1 = politics

```python
df_top_20["section"] = df_top_20["section"].apply(lambda s: s.lower())
df_top_20["is_politics"] = 1 * (df_top_20.section == "politics")
df_top_20.is_politics.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_top_20.article, df_top_20.is_politics, random_state=4567, test_size=0.2, stratify=df_top_20.is_politics)
```

```python
val_accs = []
train_accs = []
cm = []
n_ests = [10, 50, 100, 150, 200, 250]

tfidf_train = TfidfVectorizer(tokenizer=spacy_tokenizer, min_df=5, max_df=0.7)
X_train_tr = tfidf_train.fit_transform(X_train)
X_test_tr = tfidf_train.transform(X_test)

for n in n_ests:                 
    rfc = RandomForestClassifier(n_estimators=n, random_state=2375)

    rfc.fit(X_train_tr, y_train)
    y_pred_train = rfc.predict(X_train_tr)
    train_accs.append(accuracy_score(y_train, y_pred_train))

    y_pred = rfc.predict(X_test_tr)
    val_accs.append(accuracy_score(y_test, y_pred))
    cm.append(confusion_matrix(y_test, y_pred))
```

```python
plt.figure(figsize=(11,7))

plt.title("Random forest politics classifier")
plt.xlabel("Num estimators")
plt.ylabel("accuracy")

plt.plot(n_ests, val_accs, marker="o", label="validation accuracy")
plt.plot(n_ests, train_accs, marker="o", label="training accuracy")

plt.legend(fontsize=20)

plt.tight_layout()
#plt.savefig("figures/RFC_accuracy.png", facecolor='w')
```

```python
fig, axs = plt.subplots(2,3, figsize=(20, 15))

for i, n in enumerate(n_ests):
    r = i // 3
    c = i % 3
    
    axs[r, c].set_title("num est: %d" % n)
    #axs[r, c].imshow(cm[i])
    ConfusionMatrixDisplay(confusion_matrix=cm[i]).plot(ax=axs[r,c], colorbar=False)
    
plt.tight_layout()
#plt.savefig("figures/RFC_conf_mat.png", facecolor='w')
```

# Non-negative matrix factorization (NMF)

Non-negative matrix factorization (NMF) is an algorithm for approximating a matrix $V$ with a low rank matrix $WH$ so that $V \approx W H$.
Typically the quality of the approximation is measured by using the Frobenius norm as a measure of the error.
When the error is instead measured by the Kullback–Leibler divergence, NMF is identical to the probabilistic latent semantic analysis (PLSA).

NMF is commonly used in topic modeling since the non-negative entries of all matrices make it easy to interpret. Pictorially the procedure is shown below (Adapted from Blueprints for Text Analytics Using Python: Machine Learning-Based Solutions for Common Real World (NLP) Applications).

```python
from IPython.display import Image
Image("nmf_decomp.png")
```

For a given document-word matrix $V$ where the documents are the rows and the columns correspond to the frequency of words appearing in each document (e.g. measured using TF-IDF), the NMF decomposition results in the matrices $W$ and $H$. The rows of $W$ correspond to the features in each document and the columns of $H$ correspond to the words in each feature. This lends to the interpretation of each of the features as a "topic". After factorization, we can assign a topic to each document based on the largest value in each row of $W$.

```python
nmf_all_politics = NMF(n_components=15, random_state=43, alpha_W=0.0, alpha_H="same")
W_mat = nmf_all_politics.fit_transform(vectors_all_politics)
H_mat = nmf_all_politics.components_
```

```python
df_all_politics["topic"] = np.argmax(W_mat, axis=1)
```

```python
#df_all_politics.to_csv("df_all_politics.csv")
```

There appears to be a spurious topic 0 but the others can be related to issues from 2018.

```python
fig, axs = plt.subplots(5,3, figsize=(21,20))

words = tfidf_all_politics.get_feature_names_out()
for topic, weights in enumerate(H_mat):
        largest = weights.argsort()[::-1]
        d = {}
        for i in range(20):
            d[words[largest[i]]] = weights[largest[i]]

        wocl = WordCloud(background_color="white")
        wocl.generate_from_frequencies(d)
        
        axs[topic//3, topic%3].set_title("Topic %d" % topic)
        axs[topic//3, topic%3].imshow(wocl, aspect='auto', interpolation='bilinear')
        axs[topic//3, topic%3].axis("off")
```

```python
# https://matplotlib.org/3.5.0/gallery/images_contours_and_fields/image_annotated_heatmap.html
def heatmap(data, row_labels, col_labels, ax=None,
            cbar_kw={}, cbarlabel="", **kwargs):

    if not ax:
        ax = plt.gca()

    # Plot the heatmap
    im = ax.imshow(data, **kwargs)

    # Create colorbar
    cbar = ax.figure.colorbar(im, ax=ax, **cbar_kw)
    cbar.ax.set_ylabel(cbarlabel, rotation=-90, va="bottom")

    # Show all ticks and label them with the respective list entries.
    ax.set_xticks(np.arange(data.shape[1]), labels=col_labels)
    ax.set_yticks(np.arange(data.shape[0]), labels=row_labels)

    # Let the horizontal axes labeling appear on top.
    ax.tick_params(top=True, bottom=False,
                   labeltop=True, labelbottom=False)

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=-30, ha="right",
             rotation_mode="anchor")

    # Turn spines off and create white grid.
    ax.spines[:].set_visible(False)

    ax.set_xticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.grid(which="minor", color="w", linestyle='-', linewidth=3)
    ax.tick_params(which="minor", bottom=False, left=False)

    return im, cbar


def annotate_heatmap(im, data=None, valfmt="{x:.2f}",
                     textcolors=("black", "white"),
                     threshold=None, **textkw):

    if not isinstance(data, (list, np.ndarray)):
        data = im.get_array()

    # Normalize the threshold to the images color range.
    if threshold is not None:
        threshold = im.norm(threshold)
    else:
        threshold = im.norm(data.max())/2.

    # Set default alignment to center, but allow it to be
    # overwritten by textkw.
    kw = dict(horizontalalignment="center",
              verticalalignment="center")
    kw.update(textkw)

    # Get the formatter in case a string is supplied
    if isinstance(valfmt, str):
        valfmt = mpl.ticker.StrMethodFormatter(valfmt)

    # Loop over the data and create a `Text` for each "pixel".
    # Change the text's color depending on the data.
    texts = []
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            kw.update(color=textcolors[int(im.norm(data[i, j]) > threshold)])
            text = im.axes.text(j, i, valfmt(data[i, j], None), **kw)
            texts.append(text)

    return texts
```

# LSA (Latent Semantic Analysis)


LSA is based on Truncated SVD (Truncated Singular Value Decomposiiton). Truncated SVD is another algorithm for approximating a matrix $V$ with a lower rank matrix $U_k \Sigma_k V^T_k$. It is a well known fact that any real $m \times n$ matrix $V$ can be decomposed as $V=U \Sigma V^T$, where $U$ is an orthogonal $m \times m$ matrix, $V^T$ is the transpose of an orthogonal $n \times n$ matrix $V$, and $\Sigma$ is an $m \times n$ diagonal matrix with nonnegative real entries. The entries in the diagonal matrix $\Sigma$ are known as singular values.

Truncated Singular Value Decomposition considers only the $k$ largest singular values to construct $\Sigma_k$. As a result, some of the columns in $U$ and some of the rows in $V^T$ are discarded to define $U_k$ and $V^T_k$.

Although the entries in the matrices from the truncated SVD might be negative, the truncated SVD can still be easily interpreted. Pictorially the procedure is shown below (Adapted from Blueprints for Text Analytics Using Python: Machine Learning-Based Solutions for Common Real World (NLP) Applications). The subindices $k$ have been dropped for clarity.

For a given document-word matrix $V$ where the documents are the rows and the columns correspond to the frequency of words appearing in each document (e.g. measured using TF-IDF), the NMF decomposition results in the matrices $U$ and $V^T$. The rows of $U$ correspond to the features in each document and the columns of $V^T$ correspond to the words in each feature. This lends to the interpretation of each of the features as a "topic". After factorization, we can assign a topic to each document based on the largest value in each row of $U$.

## Myopicity
We define the _weekly myopicity_ of a publication as the ratio of the number of stories in a week devoted to the publication's most popular topic (in the week), to the total number of stories.

(The data used here is a saved copy of the classified stories.)

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
from matplotlib import pyplot as plt
sns.set_style('darkgrid')
mpl.rcParams['figure.dpi'] = 150
```

```python
# Load the imputed data
political_data = pd.read_csv("/media/glebzh/Backups/Erdos/csv_and_tfid_mat/df_all_politics.csv")
political_data['date'] = pd.to_datetime(political_data['date'])
```

```python
political_data['date'] = pd.to_datetime(political_data['date'])
political_data['n_week'] = political_data['date'].dt.isocalendar().week
```

```python
def percent_top_story(df_in):
    topic_count = df_in['topic'].value_counts()
    top_topic_total = topic_count.iloc[0]
    rest_topics_total = topic_count.iloc[1:].sum()
    
    if len(df_in)==0:
        data = [[0,0,0]]
    else:
        tot_topics = top_topic_total+rest_topics_total
        data = [[top_topic_total, rest_topics_total, top_topic_total/tot_topics]]
    
    columns = ['top_story', 'other_stories', 'top_percent']
    
    return pd.DataFrame(columns=columns, data=data)
```

```python
top_stories = political_data.groupby(['n_week', 'publication']).apply(
    percent_top_story
).reset_index()

top_stories.groupby('publication').apply(
    lambda x:x[['top_percent']].reset_index()[['top_percent']].T
)

time_series = top_stories.groupby('publication').apply(lambda x:x[['top_percent']].reset_index()[['top_percent']].T)
```

```python
time_series = time_series.droplevel(level=1)
```

```python
# These have very few stories, so we drop them.
pubs_to_drop = [
    'Economist',
    'Gizmodo',
    'Hyperallergic',
    'New Republic',
    'TechCrunch',
    'Washington Post',
    'Wired',
    'The Verge',
]
```

```python
time_series = time_series[~time_series.index.isin(pubs_to_drop)]
top_stories = top_stories[~top_stories.publication.isin(pubs_to_drop)]
```

```python
top_stories_avg = pd.DataFrame(
    top_stories.groupby('publication')['top_percent'].agg(['mean', 'std']).sort_values('mean')[:-4]
).reset_index()
```

```python
order = top_stories[['publication', 'top_percent']].groupby('publication').mean().sort_values('top_percent').index
```

```python
plt.figure(figsize=(8,5),dpi=150)
ax = sns.barplot(data=top_stories, y='publication', x='top_percent', order=order)
ax.set_xlabel("fraction of weekly articles dedicated to dominating topic (myopicity)")
ax.set_title("Articles dedicated to dominating topic")
plt.tight_layout()
plt.savefig("stories_fraction.png", dpi=200)
```
